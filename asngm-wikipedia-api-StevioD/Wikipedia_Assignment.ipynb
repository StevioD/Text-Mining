See the README for the full description. In this assignment we'll pull all the names for people born in the 150 years ending in 2015. 

```python
# Some code to get you started
import requests
from collections import Counter
from collections import defaultdict
from pprint import pprint

wikipedia_api_url = "https://en.wikipedia.org/w/api.php"

req = {'action':'query',
          'format':'json',
          'list':'categorymembers',
          'cmlimit':10}
```

```python
# getting a list of year between 1865-2016 for the loop reference 
req = {'action':'query',
          'format':'json',
          'list':'categorymembers',
          'cmlimit':10}

year = 1865
year_list = []

while year < 2016:
    year_list.append(year)
    year += 1
```

```python
# Create a dictionary and loop through each year, scrape through each page on wiki for each year 

names_by_year = {}

for year in year_list:
    
    req = {'action':'query',
          'format':'json',
          'list':'categorymembers',
          'cmlimit':500,
          'cmtitle':'Category:' + str(year) + '_births'}
    
    last_continue = {}
    iteration = 1
    pages = 0
    
    # Modify it with the values returned in the 'continue' section of the last result.
    req.update(last_continue)
    
    # Call API
    result = requests.get('https://en.wikipedia.org/w/api.php', params=req).json() 
    
    pages += 1
    
    # Grab the names
    names_list = []
    
    for item in result['query']['categorymembers'] :
        names_list.append(item['title'])
        
    names_by_year[year] = names_list
        
    iteration += 1
    
       # Can we get out?
    if 'continue' not in result :
        break
    else :
        last_continue = result['continue']
```

```python
birth_year = []
names = []

year = 1865

while year < 2007:
    for name in names_by_year[year]:
        birth_year.append(year)
        names.append(name)
        year += 1
```

```python
# print the list
print(len(names))
print(len(birth_year))
```

```python
import csv
with open('names_by_year.csv', 'w', encoding = 'utf-8') as outfile:
    writer = csv.writer(outfile, delimiter='\t')
    writer.writerows(zip(birth_year,names))
```
